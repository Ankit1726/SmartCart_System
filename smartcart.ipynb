- Data Preprocessing

```python
## Handling Missing Values
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
## Feature Engineering
df.columns
```

```python
## Adding Age Feature
df["Age"] = 2026-df["Year_Birth"]
```

```python
df.head()
```

```python
## Customer Joining Date
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],dayfirst=True)

refrence_date = df["Dt_Customer"].max()
df["Customer_Tenure_Days"] = (refrence_date-df["Dt_Customer"]).dt.days
```

```python
df.head()
```

```python
df.columns
```

```python
## Spending
df["Total_Spending"] = df["MntWines"] +df["MntFruits"] + df["MntMeatProducts"] +df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
```

```python
## Children
df["Total_Children"] = df["Kidhome"] + df["Teenhome"]
```

```python
## Education
df["Education"] = df["Education"].replace({
     "Basic":"Undergraduate",
     "2n Cycle":"Undergraduate",
     "Graduation":"Graduate",
     "Master":"Postgraduate",
     "PhD":"Postgraduate"
})

df["Education"].value_counts()
```

```python
## Marital Status
df["Living_With"] = df["Marital_Status"].replace({
     "Married":"Partner",
     "Together":"Partner",
     "Single":"Alone",
     "Widow":"Alone",
     "Divorced":"Alone",
     "Absurd":"Alone",
     "YOLO":"Alone"
})

df["Living_With"].value_counts()
```

```python
df.head()
```

```python
## Columns
cols = ["ID","Year_Birth","Marital_Status","Kidhome","Teenhome","Dt_Customer"]
spending_cols = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds"]

cols_to_drop = cols + spending_cols
df_cleaned = df.drop(columns=cols_to_drop)
df_cleaned.shape
```

```python
df_cleaned.head()
```

```python
## Finding Outliers
cols = ["Income","Recency","Response","Age","Total_Spending","Total_Children"]

## relative plots of some features (pair plot)
sns.pairplot(df_cleaned[cols])
```

```python
## Remove The Outliers
print(f'Data Size With Outliers:',len(df_cleaned))
df_cleaned = df_cleaned[(df_cleaned["Age"] < 90)]
df_cleaned = df_cleaned[(df_cleaned["Income"] < 600_000)]

print(f'Data Size WithOut Outliers:',len(df_cleaned))
```

```python
## HeatMap
corr = df_cleaned.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
     corr,
     annot=True,
     annot_kws={"size":7},
     cmap="coolwarm"
)
```

- Encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
cat_cols = ["Education","Living_With"]
enc_cols = ohe.fit_transform(df_cleaned[cat_cols])
enc_df = pd.DataFrame(enc_cols.toarray(),columns=ohe.get_feature_names_out(cat_cols),index=df_cleaned.index)
```

```python
df_encoded = pd.concat([df_cleaned.drop(columns=cat_cols),enc_df],axis=1)
```

```python
df_encoded.shape
```

```python
df_encoded.head()
```

- Scaling Our Data

```python
from sklearn.preprocessing import StandardScaler
X = df_encoded
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

```python
## Visualizing Our Data
x_scaled.shape
```

```python
## 2D Visualization
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
## Plot Graph
plt.scatter(x_pca[:,0],x_pca[:,1])
```

```python
pca.explained_variance_ratio_
```

```python
## 3D Visualization
pca = PCA(n_components=3)
x_pca = pca.fit_transform(x_scaled)
pca.explained_variance_ratio_
```

```python
## Plot Graph
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111,projection="3d")
ax.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2])

ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3D Projections")
```

- Analyze K values

```python
## Elbow Method
from sklearn.cluster import KMeans
from kneed import KneeLocator

wcss = []
for k in  range(1,11):
     kmeans = KMeans(n_clusters=k,random_state=42)
     kmeans.fit_predict(x_pca)
     wcss.append(kmeans.inertia_)
```

```python
knee = KneeLocator(range(1,11),wcss,curve="convex",direction="decreasing")
optimal_k = knee.elbow
```

```python
print(f"Best K:{optimal_k}")
```

```python
## Plot Curve
plt.plot(range(1,11),wcss,marker="o")
plt.xlabel("K_Values")
plt.ylabel("WCSS")
```

```python
## Silhoute Score
from sklearn.metrics import silhouette_score
scores = []
for k in  range(2,11):
     kmeans = KMeans(n_clusters=k,random_state=42)
     labels = kmeans.fit_predict(x_pca)
     score = silhouette_score(x_pca,labels)
     scores.append(score)

## Plot
plt.plot(range(2,11),scores,marker="o")
plt.xlabel("K_Values")
plt.ylabel("Silhoutte Score")
```

```python
## Combined Plot
k_range = range(2,11)
fig,ax1 = plt.subplots(figsize=(8,6))
ax1.plot(k_range,wcss[:len(k_range)],marker="o",color="green")
ax1.set_label("K_values")
ax1.set_ylabel("WCSS")

ax2 = ax1.twinx()
ax2.plot(k_range,scores[:len(k_range)],marker="x",color="red",linestyle="--")
ax2.set_ylabel("Silhoutte Score")
```

- Clustering

```python
# KMeans
kmeans = KMeans(n_clusters=4,random_state=42)
labels_kmeans = kmeans.fit_predict(x_pca)

## 3D Plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection="3d")
ax.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2],c=labels_kmeans)
```

```python
## Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering

agg_clf = AgglomerativeClustering(n_clusters=4,linkage="ward")
labels_agg = agg_clf.fit_predict(x_pca)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection="3d")
ax.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2],c=labels_agg)
```

- Characterization of clustering

```python
df_cleaned.drop("labels",axis=1)
X["cluster"] = labels_agg
X.head()
```

```python
pal = ["red","blue","yellow","green"]
sns.countplot(x=X["cluster"],palette=pal,hue=X["cluster"])
```

```python
## Income & Spending Patterns
sns.scatterplot(x=X["Total_Spending"],y=X["Income"],hue=X["cluster"],palette=pal)
```

- Cluster Summary

```python
cluster_summary = X.groupby("cluster").mean()
print(cluster_summary)
```

- Final Trained Model

```python
# One-hot encoding
df_encoded = pd.get_dummies(
    df,
    columns=["Education", "Living_With"],
    drop_first=False
)
```

```python
FINAL_FEATURES = [
    'Income', 'Recency', 'NumDealsPurchases', 'NumWebPurchases',
    'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
    'Complain', 'Response', 'Age', 'Customer_Tenure_Days',
    'Total_Spending', 'Total_Children',
    'Education_Graduate', 'Education_Postgraduate',
    'Education_Undergraduate',
    'Living_With_Alone', 'Living_With_Partner'
]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pickle

X = df_encoded[FINAL_FEATURES]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_scaled)

pickle.dump(kmeans, open("model.pkl", "wb"))
pickle.dump(scaler, open("scaler.pkl", "wb"))
```
